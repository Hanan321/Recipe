```python
print("Choose a city:")
print("1. Riyadh")
print("2. Cairo")
print("3. Amman")
print("4. Rabat")
print("5. Tripoli")
print("6. Khartoum")
```

```python
choice = input("Enter the number of your choice: ").strip()
```

```python
city_options = {
    "1": "Riyadh",
    "2": "Cairo",
    "3": "Amman",
    "4": "Rabat",
    "5": "Tripoli",
    "6": "Khartoum"
}

city = city_options.get(choice)

if city:
    print("You selected:", city)
else:
    print("Invalid choice")
```

```python
city_dishes = {
    "Riyadh": "Chicken Kabsa",
    "Cairo": "Coming Soon",
    "Amman": "Coming Soon",
    "Rabat": "Coming Soon",
    "Tripoli": "Coming Soon",
    "Khartoum": "Coming Soon"
}

dish = city_dishes.get(city)
print("Dish:", dish)
```

```python
recipes = {
    "Chicken Kabsa": {
        "ingredients": [
            "rice", "chicken", "tomatoes", "onion",
            "garlic", "spices", "carrots"
        ],
        "steps": [
            "Cook chicken with spices",
            "Add tomatoes and broth",
            "Add rice and simmer",
            "Serve hot"
        ],
        "cost": "$12 - $18",
        "plating": "A large platter of rice topped with golden chicken",
        "image_prompt": "Saudi chicken kabsa, realistic food photography"
    }
}
```

```python
recipe = recipes.get(dish)

if recipe:
    print("\n=== RESULT ===")
    print("City:", city)
    print("Dish:", dish)

    print("\nIngredients:")
    for i in recipe["ingredients"]:
        print("-", i)

    print("\nSteps:")
    for i, step in enumerate(recipe["steps"], 1):
        print(f"{i}. {step}")

    print("\nCost:", recipe["cost"])
    print("\nPlating:", recipe["plating"])
    print("\nImage Prompt:", recipe["image_prompt"])

else:
    print("Recipe not added yet.")
```
