```python
from openai import OpenAI

client = OpenAI()

city_options = {
    "1": "Riyadh",
    "2": "Cairo",
    "3": "Amman",
    "4": "Rabat",
    "5": "Tripoli",
    "6": "Khartoum"
}

city_dishes = {
    "Riyadh": "Chicken Kabsa",
    "Cairo": "Koshari",
    "Amman": "Mansaf",
    "Rabat": "Chicken Tagine",
    "Tripoli": "Bazeen",
    "Khartoum": "Mulah"
}

print("Arabic AI Chef")
print("1. Riyadh")
print("2. Cairo")
print("3. Amman")
print("4. Rabat")
print("5. Tripoli")
print("6. Khartoum")

choice = input("Enter number: ").strip()
city = city_options.get(choice)

if not city:
    print("Invalid choice")
else:
    dish = city_dishes[city]

    prompt = f"""
You are a professional culinary AI assistant.

City: {city}
Dish: {dish}

Generate a clean and structured answer with these exact sections:

Ingredients:
- list 6 to 8 items

Recipe Steps:
- write 4 to 6 simple steps

Estimated Cost:
- give a realistic price range in USD

Plating Description:
- describe how the dish looks when served

Image Prompt:
- write a realistic food photography prompt
"""

    response = client.responses.create(
        model="gpt-5.4",
        input=prompt
    )

    print("\n" + "=" * 60)
    print("ARABIC AI CHEF")
    print("=" * 60)
    print(f"City: {city}")
    print(f"Dish: {dish}")
    print("-" * 60)
    print(response.output_text)
    print("=" * 60)
```
