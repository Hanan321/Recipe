```python
import os
from getpass import getpass

os.environ["OPENAI_API_KEY"] = getpass("Paste your API key: ")
```

```python
from openai import OpenAI

client = OpenAI()

print("\n🍽️ Arabic AI Chef")
print("Welcome! Let's explore Arabic cuisine with AI ideas.")

arab_countries = [
    "Algeria", "Bahrain", "Comoros", "Djibouti", "Egypt",
    "Iraq", "Jordan", "Kuwait", "Lebanon", "Libya",
    "Mauritania", "Morocco", "Oman", "Palestine", "Qatar",
    "Saudi Arabia", "Somalia", "Sudan", "Syria",
    "Tunisia", "United Arab Emirates", "Yemen"
]

normalized_countries = {country.lower(): country for country in arab_countries}


def generate_recipe(country: str) -> str:
    prompt = f"""
You are a professional culinary AI assistant.

The selected country is: {country}

Choose ONE well-known traditional dish from this country and generate the answer using this exact format:

Dish Name:
- write the name of one popular traditional dish

Plating Description:
- describe how the dish looks when served

Ingredients:
- list 6 to 8 items

Recipe Steps:
1. write 4 to 6 simple steps

Keep the answer culturally appropriate, realistic, and beginner-friendly.
"""

    response = client.responses.create(
        model="gpt-5.4",
        input=prompt
    )
    return response.output_text


def show_intro() -> None:
    print("\n🍽️ Arabic AI Chef")
    print("Enter an Arabic country to generate a traditional dish.")
    print("Type 'help' to see supported countries or 'q' to quit.")


while True:
    show_intro()
    user_input = input("\nEnter a country name: ").strip().lower()

    if user_input == "q":
        print("Goodbye!")
        break

    if user_input == "help":
        print("\n🌍 Supported Arab Countries:")
        print(", ".join(arab_countries))
        continue

    if user_input not in normalized_countries:
        print("\n❌ Invalid country. Type 'help' to see the supported list.")
        continue

    country = normalized_countries[user_input]
    print("🤖 AI is selecting a traditional dish...\n")
    print("\nGenerating AI recipe...\n")
    output_text = generate_recipe(country)

    print("\n" + "=" * 60)
    print("🍽️ ARABIC AI CHEF")
    print("=" * 60)
    print(f"🌍 Country: {country}")
    print("-" * 60)
    print(output_text)
    print("=" * 60)

    again = input("\n🔁 Do you want to choose another country? (y/n): ").strip().lower()

    if again != "y":
        print("\nThank you for using Arabic AI Chef 👋")
        break
```
